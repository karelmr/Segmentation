```python
country_counts = data['Nationality'].value_counts()

# Selecionando apenas os 20 principais países
top_20_countries = country_counts[:20]

# Criando um gráfico de barras
plt.figure(figsize=(10,5))
top_20_countries.plot(kind='bar')
plt.title('Distribuição de Países')
plt.xlabel('País')
plt.ylabel('Contagem')
plt.show()
```

```python
# Contando o número de idades menores que 0
count = (data['Age'] < 18).sum()

print("Número de idades menores que 0:", count)
```

```python
# Criando um boxplot da idade
plt.figure(figsize=(3,2))
plt.boxplot(data['Age'].dropna())
plt.title('Boxplot da Idade')
plt.xlabel('Idade')
plt.show()
```

```python
# Mantendo apenas as linhas onde 'Age' está entre 18 e 90
data2 = data[(data['Age'] >= 18) & (data['Age'] <= 90)]
data2.shape
```

```python
# Criando um boxplot da idade
plt.figure(figsize=(3,2))
plt.boxplot(data2['Age'].dropna())
plt.title('Boxplot da Idade')
plt.xlabel('Idade')
plt.show()
```
